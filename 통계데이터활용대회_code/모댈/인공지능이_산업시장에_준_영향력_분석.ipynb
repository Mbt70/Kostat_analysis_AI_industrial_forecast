```python
import pandas as pd
import numpy as np

data1 = pd.read_csv('year_mapped.csv', index_col=0,encoding='cp949')
data2 = pd.read_csv('.csv', encoding='cp949', index_col=0)
data3 = pd.read_csv("__.csv")
```

```python
for col in data1.columns:
  if data1[col].dtype == 'object':
    data1[col] = data1[col].apply(lambda x: float(x.strip().replace(',', '')) if pd.notnull(x) else x)

data3 = data3[list(range(0, 63, 4))]
data3.columns = data3.iloc[0]
data3 = data3.iloc[2:]
data3 = data3.iloc[14:]
data3
```

```python
from pycaret.time_series import *

for target_col in dependent_df.columns:
  train_set = pd.concat([independent_df, dependent_df[target_col]], axis=1)
  train_set.index = pd.to_datetime(train_set.index)

  before_2016 = TSForecastingExperiment()

  before_2016.setup(
      data=train_set.iloc[:108],
      target=target_col,
      scale_exogenous='robust',
      transform_exogenous='box-cox'
  )

  random_forest = before_2016.create_model('croston')
  print(target_col)
  plt.plot(train_set[target_col])
  plt.plot(before_2016.predict_model(random_forest, fh=61)['y_pred'])
  plt.show()
```
