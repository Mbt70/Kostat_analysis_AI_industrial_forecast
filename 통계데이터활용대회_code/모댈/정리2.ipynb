```python
from pycaret.time_series import *
from boruta import BorutaPy
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

exclude_models = [
    "auto_arima",
    "bats",
    "tbats",
    "lar_cds_dt",
    "par_cds_dt",
]

def train(independent_df, dependent_df, target_column):
  independent_df_new = independent_df.copy().iloc[:248]
  for col in independent_df_new.columns:
    for lag in [3, 12, 24]:
      independent_df_new[f'{col}_lag_{lag}'] = independent_df_new[col].shift(lag)

  # columns_with_missing = independent_df_new.columns[independent_df_new.isnull().any()].tolist()
  # df_to_impute = independent_df_new[columns_with_missing].copy()
  # rf_imputer = IterativeImputer(
  #   estimator=RandomForestRegressor(n_estimators=100, random_state=42),
  #   max_iter=10,
  #   random_state=42
  # )
  # df_imputed = pd.DataFrame(
  #   rf_imputer.fit_transform(df_to_impute),
  #   columns=df_to_impute.columns,
  #   index=df_to_impute.index
  # )

  # independent_df_new[columns_with_missing] = df_imputed

  train_set = pd.concat([independent_df_new.copy().iloc[:248], dependent_df[target_column].iloc[:248]], axis=1)
  predicter = setup(data=train_set, target=target_column, fh=12, scale_exogenous='robust', session_id=42, use_gpu=True, numeric_imputation_exogenous='ffill')

  # 보루타로 중요 외생변수 선택
  rf = RandomForestRegressor(n_estimators='auto', random_state=42)
  boruta = BorutaPy(rf, n_estimators='auto', random_state=42)

  X_train = predicter.X_transformed.copy()
  y_train = dependent1.iloc[:248][target_column]

  boruta.fit(X_train.values, y_train.values)
  selected_features = X_train.columns[boruta.support_].to_list()

  train_set = train_set[[target_column]+selected_features]

  # 선택된 외생변수 각각에 대한 시계열 예측 수행
  exog_exps = []
  exog_models = []

  for exog_var in selected_features:
    exog_exp = TSForecastingExperiment()

    exog_exp.setup(
      data=train_set[exog_var],
      target=exog_var,
      fh=48,
      numeric_imputation_target="ffill",
      numeric_imputation_exogenous="ffill",
      session_id=42,
      verbose=False,
    )

    best = exog_exp.compare_models(
      sort="mase",
      include=[
          "arima",
          "ets",
          "exp_smooth",
          "theta",
          "lightgbm_cds_dt",
      ],
    )

    final_exog_model = exog_exp.finalize_model(best)

    exog_exps.append(exog_exp)
    exog_models.append(final_exog_model)

  future_exog = [
    exog_exp.predict_model(exog_model) for exog_exp, exog_model in zip(exog_exps, exog_models)
  ]

  future_exog = pd.concat(future_exog, axis=1)
  future_exog.columns = selected_features

  print(future_exog)

  # 모델 학습
  exp_auto = TSForecastingExperiment()

  exp_auto.setup(
      data=train_set,
      target=target_column,
      fh=48,
      scale_exogenous='robust',
      enforce_exogenous=True,
      session_id=42,
  )

  best = exp_auto.compare_models(sort="mase")
  final_best = exp_auto.finalize_model(best)

  result = exp_auto.predict_model(final_best, X=future_exog)

  return result
```

```python
test = train(merged_independent, dependent1, '전직종_구직_비율')
```
