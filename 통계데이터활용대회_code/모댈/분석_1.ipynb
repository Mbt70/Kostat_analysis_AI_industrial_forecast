```python
import pandas as pd
import numpy as np

# 1. 데이터 확인
print("keco shape:", keco.shape)
print("keco index:", keco.index)
print("keco columns:", keco.columns)

# 2. 2008년과 2024년(또는 가장 최근 연도) 데이터 추출
data_2008 = keco.loc['2008'] if '2008' in keco.index else None
data_2024 = keco.loc['2024'] if '2024' in keco.index else keco.iloc[-1]

print("\ndata_2008 shape:", data_2008.shape if data_2008 is not None else "Not found")
print("data_2024 shape:", data_2024.shape)

# 3. 변화율 계산 (NaN 및 inf 값 처리)
if data_2008 is not None and not data_2008.empty and not data_2024.empty:
    change_rate = ((data_2024 - data_2008) / data_2008 * 100)
    change_rate = change_rate.replace([np.inf, -np.inf], np.nan).dropna()
    
    print("\nchange_rate shape:", change_rate.shape)
    print("change_rate non-null count:", change_rate.count())
    
    if not change_rate.empty:
        # 4. 상위 5개 및 하위 5개 변화율 칼럼 식별
        top_5_increase = change_rate.nlargest(5)
        top_5_decrease = change_rate.nsmallest(5)
        
        print("\nTop 5 columns with the largest increase in change rate:")
        print(top_5_increase)
        print("\nTop 5 columns with the largest decrease in change rate:")
        print(top_5_decrease)
    else:
        print("\nNo valid data in change_rate after filtering")
else:
    print("\nUnable to calculate change rate due to missing data")
```
