```python
#전국사업체 조사 , 기업활동조사 결합
df=pd.read_csv("산업별_사업체수.csv",encoding='cp949')
# Separate '계' and others
df_total = df[df['종사자규모별'] == '계'].set_index(['산업별', '종사자규모별'])
df_others = df[df['종사자규모별'] != '계'].groupby('산업별').sum(numeric_only=True)

# Calculate the difference
df_diff = df_total - df_others


df_diff
```

```python

diff_2020 = df_diff.loc[:,'2020_1'] - df_diff.loc[:,'2020']
diff_2020

# Adding the difference to the appropriate columns (up to '2020')
df_diff.loc[:, :'2020'] = df_diff.loc[:, :'2020'].add(diff_2020, axis=0)
df_result=df_diff.drop('2020_1',axis=1)
df_result=df_result.transpose()
df_result.columns = df_result.columns.droplevel(1)
df_result=df_result.drop('공공행정, 국방 및 사회보장 행정(84)',axis=1)
df_result

indust_consume.columns=df_result.columns
indust_consume.index=df_result.index

df_result = df_result.apply(pd.to_numeric, errors='coerce')
indust_consume.apply(pd.to_numeric, errors='coerce')
indust_sum=indust_consume.copy()*df_result.copy()
indust_sum=indust_sum.iloc[2:,:]

display(indust_sum)

def convert_date(date_str):
    year=str(date_str)
    month = '01'
    return pd.to_datetime(f'{year}-{month}-01')

indust_sum.index=indust_sum.index.map(convert_date)
date_range = pd.date_range(start='2008-01-01', end='2022-01-01', freq='MS')

indust_monthly = pd.DataFrame(index=date_range, columns=indust_sum.columns)

for date, row in indust_sum.iterrows():
    indust_monthly.loc[date] = row

# 큐빅 스플라인 보간법 적용 함수
def cubic_spline_interpolation(series):
    series = pd.to_numeric(series, errors='coerce')
    non_nan = series.dropna()
    if len(non_nan) < 4:
        return series.interpolate(method='linear')
    x = np.arange(len(series))
    y = series.values
    mask = ~np.isnan(y)
    if np.sum(mask) < 4:
        return series.interpolate(method='linear')
    try:
        f = interpolate.interp1d(x[mask], y[mask], kind='cubic', fill_value='extrapolate')
        return pd.Series(f(x), index=series.index)
    except:
        return series.interpolate(method='polynomial', order=3)

# 각 열에 대해 큐빅 스플라인 보간 적용
for column in indust_monthly.columns:
    indust_monthly[column] = cubic_spline_interpolation(indust_monthly[column])


indust_monthly.to_csv("산업별_매출_진짜최종.csv")
# Plot the first column
plt.figure(figsize=(10, 6))
indust_monthly.iloc[:, 0].plot(title='First Column Cubic Spline Interpolation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
